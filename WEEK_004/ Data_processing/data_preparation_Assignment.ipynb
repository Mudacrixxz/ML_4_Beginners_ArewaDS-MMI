```python
import pandas as pd
import seaborn as sns
pumpkins = pd.read_csv('../WEEK_004/US-pumpkins.csv')
import matplotlib.pyplot as plt
pumpkins.head()
```

```python
# Plotting the data
plt.scatter('Low Price','Mostly Low',data=pumpkins)
plt.xlabel('Low Prices')
plt.ylabel('Mostly Low')
plt.title('Relationship between Low and Mostly Low Prices of US Pumpkins')

# Display the plot
plt.show()
```

```python
# Plotting the data
plt.plot('Mostly Low','High Price',data=pumpkins)
plt.xlabel('Mostly Low')
plt.ylabel('High Prices')
plt.title('Relationship between Mostly Low and High Prices of US Pumpkins')

# Display the plot
plt.show()
```

```python
# using seaborn
sns.histplot(data=pumpkins,x='Low Price', y= 'High Price')
plt.title('Relationship between High and Low Prices')
plt.xlabel('Low Price')
plt.ylabel('High Price')
plt.show()
```
