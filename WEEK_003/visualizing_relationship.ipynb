```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
honey = pd.read_csv('honey.csv')
honey.head(30)
```

```python
honey.isnull().sum()
```

```python
sns.relplot(x='year',y='stocks',
            col='state',
            col_wrap=3,
            kind='scatter', data=honey
            )
```

```python
sns.relplot(
            x='state',y='prodvalue',
            col='year',
            col_wrap=3,
            kind='line', data=honey)
```

```python
fig, ax = plt.subplots(figsize=(12,6))
lineplot = sns.lineplot(x=honey['year'], y=honey['numcol'], data=honey, 
                        label = 'Number of bee colonies', legend=False)
sns.despine()
plt.ylabel('# colonies')
plt.title('Price vs number of bee colonies over the years');

ax2 = ax.twinx()
lineplot2 = sns.lineplot(x=honey['year'], y=honey['priceperlb'], ax=ax2, color="r", 
                         label ='Price per lb', legend=False) 
sns.despine(right=False)
plt.ylabel('price per lb')
ax.figure.legend();
```
